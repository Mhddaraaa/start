```python
import qrcode
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
text = "www.google.com"
im = qrcode.make(text)

plt.imsave('qrGoogle.jpg', im, cmap='gray')
plt.imshow(im, cmap='gray')
plt.show()
```

```python
img = cv.imread('qrGoogle.jpg', 0)

detector = cv.QRCodeDetector()
data, bbox, straight_qrcode = detector.detectAndDecode(img)
data, bbox
```
