# 1_Gaussian filter
### filter size must be at least 3 times sigma in every direction

```python
def guassianFilter(kernel_size, sigma):
    ranges = kernel_size // 2
    kernel = np.zeros((kernel_size, kernel_size))
    for i in range(kernel_size):
        x = i - ranges
        for j in range(kernel_size):
            y = j - ranges
            power = (np.square(x) + np.square(y)) / (2 * np.square(sigma))
            base = 1 / (2 * np.pi * np.square(sigma))
            kernel[i, j] = base * np.exp(- power)
            
    return kernel
```

```python
def gaussianFilterDerivative(kernel_size, sigma):
    ranges = kernel_size // 2
    kernel_x = np.zeros((kernel_size, kernel_size))
    kernel_y = np.zeros((kernel_size, kernel_size))
    for i in range(kernel_size):
        x = i - ranges
        for j in range(kernel_size):
            y = j - ranges
            base_x = - x / (2 * np.pi * np.square(np.square(sigma)))
            base_y = - y / (2 * np.pi * np.square(np.square(sigma)))
            power = (np.square(x) + np.square(y)) / (2 * np.square(sigma))

            kernel_x[i, j] = base_x * np.exp(- power)
            kernel_y[i, j] = base_y * np.exp(- power)
            
    return kernel_x, kernel_y
```

```python
def NMS(Gradient_magnitude, Gradient_direction):
    Gd_bins = 45 * np.round(Gradient_direction / 45)
    row, col = Gradient_magnitude.shape
    Gm_NMS = np.zeros((row, col))
    for i in range(1, row -1):
        for j in range(1, col - 1):
            value = Gd_bins[i, j]
            if value in [0.0, -0.0, 180., -180.]:
                a, b = Gradient_magnitude[i - 1, j], Gradient_magnitude[i + 1, j]
            elif value in [-45., 135]:
                a, b = Gradient_magnitude[i - 1, j + 1], Gradient_magnitude[i + 1, j - 1]
            elif value in [45., -135]:
                a, b = Gradient_magnitude[i - 1, j - 1], Gradient_magnitude[i + 1, j + 1]
            elif value in [-90., 90.]:
                a, b = Gradient_magnitude[i, j - 1], Gradient_magnitude[i, j + 1]
            else:
                print('Error')
                return
            
            if Gradient_magnitude[i, j] >= a and Gradient_magnitude[i, j] >= b:
                Gm_NMS[i, j] = Gradient_magnitude[i, j]
    
    return Gm_NMS
```

```python
def guassianFilter(kernel_size, sigma):
    ranges = kernel_size // 2
    kernel = np.zeros((kernel_size, kernel_size))
    for i in range(kernel_size):
        x = i - ranges
        for j in range(kernel_size):
            y = j - ranges
            power = (np.square(x) + np.square(y)) / (2 * np.square(sigma))
            base = 1 / (2 * np.pi * np.square(sigma))
            kernel[i, j] = base * np.exp(- power)
            
    return kernel
```

```python
def canny_costume(img):
    # Smooth image with gaussian filter
    img_s = cv.filter2D(img, -1, guassianFilter(11, 2))
    img_s = img_s[:, :, 0].astype(np.float64)
    
    # compute gradient magnitude and direction numericaly
    x_mask = np.array([[-1],
                       [1]])
    y_mask = x_mask.T
    
    Ix = cv.filter2D(img_s, -1, x_mask)
    Iy = cv.filter2D(img_s, -1, y_mask)

    Gradient_magnitude = np.sqrt(np.square(Ix) + np.square(Iy))
    Gradient_direction = np.rad2deg(np.arctan2(Iy, Ix))
    
    # Non-maxima supression, to know compare element in which direction
    Gd_bins = 45 * np.round(Gradient_direction / 45)
    row, col = Gradient_magnitude.shape
    Gm_NMS = np.zeros((row, col))
    for i in range(1, row -1):
        for j in range(1, col - 1):
            value = Gd_bins[i, j]
            if value in [0.0, -0.0, 180., -180.]:
                a, b = Gradient_magnitude[i - 1, j], Gradient_magnitude[i + 1, j]
            elif value in [-45., 135]:
                a, b = Gradient_magnitude[i - 1, j + 1], Gradient_magnitude[i + 1, j - 1]
            elif value in [45., -135]:
                a, b = Gradient_magnitude[i - 1, j - 1], Gradient_magnitude[i + 1, j + 1]
            elif value in [-90., 90.]:
                a, b = Gradient_magnitude[i, j - 1], Gradient_magnitude[i, j + 1]
            else:
                print('Error')
                return
            
            if Gradient_magnitude[i, j] >= a and Gradient_magnitude[i, j] >= b:
                Gm_NMS[i, j] = Gradient_magnitude[i, j]
    
    # select best threshold for edges
    L = Gm_nms.mean()
    H = L + Gm_nms.std()
    filtered_img = filters.apply_hysteresis_threshold(Gm_nms, L, H)
    
    return filtered_img
```

```python
img = cv.imread('people2.png')
img_edge_cv = cv.Canny(img, 150, 250)
img_edge_costume = canny_costume(img)
```

```python
fig = plt.figure(figsize=(12, 8))
plt.subplot(121)
plt.imshow(img_edge_cv, cmap='gray')
plt.title('Cv')
plt.subplot(122)
plt.imshow(img_edge_costume, cmap='gray')
plt.title('Costume')
plt.show()
```
