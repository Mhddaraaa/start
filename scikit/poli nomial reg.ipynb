```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(6)
m = 100
X = 6 * np.random.rand(m, 1) - 3
X = np.sort(X)

y = 0.5 * X ** 2 + X + 2 + np.random.rand(m, 1)
y = np.sort(y)

plt.style.use('fivethirtyeight')
plt.scatter(X, y)
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)
```

```python
line_reg = LinearRegression()
line_reg.fit(X_poly, y)


Y = line_reg.predict(X_poly)
line_reg.intercept_, line_reg.coef_
```

```python
# c = line_reg.intercept_
# b, a = line_reg.coef_[0, 0], line_reg.coef_[0, 1]

# Y = a * X ** 2 + b * X + c

# plt.plot(data=(X.sort(),Y.sort()))

df = pd.DataFrame(X, columns = ['X'])

df['Y'] = Y
df = df.sort_values('X')


plt.plot(df['X'], df['Y'], color='r')
plt.scatter(X, y)
df
```

```python
poly = PolynomialFeatures(degree=1, include_bias=False)
X_poly = poly.fit_transform(X)
line_reg = LinearRegression()
line_reg.fit(X_poly, y)


Y = line_reg.predict(X_poly)
plt.scatter(X, Y)
```

```python
poly = PolynomialFeatures(degree=40, include_bias=False)
X_poly = poly.fit_transform(X)
line_reg = LinearRegression()
line_reg.fit(X_poly, y)

Y2 = line_reg.predict(X_poly)

df2 = pd.DataFrame(X, columns = ['X'])

df2['Y'] = Y2
df2 = df2.sort_values('X')

plt.plot(df2['X'], df2['Y'], color='r', linewidth=2)
plt.scatter(X, y)
```
