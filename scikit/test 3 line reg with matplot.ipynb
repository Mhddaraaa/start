# Mean Absolute Error, mean Square Error,  Root mean Square Error

### MAE/MSE/RMSE

```python
from sklearn import metrics

true = [100, 50, 30, 20]
predict = [90, 50, 50, 30]
```

```python
print((10 + 0 + 20 + 10) / 4)

print(metrics.mean_absolute_error(true, predict))
```

```python
print((10 ** 2 + 0 ** 2 + 20 ** 2 + 10 ** 2) / 4)

print(metrics.mean_squared_error(true, predict))
```

```python
np.sqrt(metrics.mean_squared_error(true, predict))
```

```python
np.sqrt(metrics.mean_squared_error(y_test, y_pred))
```

# matplotlib reg


```python
import matplotlib.pyplot as plt
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize = (15, 5), sharey=True)

ax1.scatter(df['TV'].tolist(), df['Sales'].tolist())
ax2.scatter(df['Radio'].tolist(), df['Sales'].tolist(), c='g')
ax3.scatter(df['Newspaper'].tolist(), df['Sales'].tolist(), c='r')
```

```python
fig, ax = plt.subplots()
x = df['TV'].tolist()
y = df['Sales'].tolist()

ax.scatter(x, y)

m, b = np.polyfit(x, y, deg=1)

# ax.axline(xy1=(0, b), slope=m, color='r', label=f'$y = {m:.2f}x {b:+.2f}$')
ax.plot(x , m * (np.array(x)) + b, color='k')
```

```python
# plt.style.use('seaborn')

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize = (15, 5), sharey=True)

x1 = df['TV'].tolist()
x2 = df['Radio'].tolist()
x3 = df['Newspaper'].tolist()
y = df['Sales'].tolist() 

ax1.scatter(x1, y)
ax2.scatter(x2, y, c='g')
ax3.scatter(x3, y, c='r')

m1, b1 = np.polyfit(x1, y, deg=1)
m2, b2 = np.polyfit(x2, y, deg=1)
m3, b3 = np.polyfit(x2, y, deg=1)

ax1.plot(x1 , m1 * (np.array(x1)) + b1, color='k')
ax2.plot(x2 , m2 * (np.array(x2)) + b2, color='k')
ax3.plot(x3 , m3 * (np.array(x3)) + b3, color='k')

ax1.set_xlabel('TV spending money')
ax2.set_xlabel('Radio spending money')
ax3.set_xlabel('News paper spending money')
```
