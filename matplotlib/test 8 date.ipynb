```python
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_date
import pandas as pd
from datetime import datetime,timedelta


plt.style.use('fivethirtyeight')

date = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]
y = [0, 1, 3, 5, 4, 5, 2]

plt.plot_date(date, y, linestyle='-', linewidth=2)

plt.gcf().autofmt_xdate()

date_format = mpl_date.DateFormatter('%b, %d %Y')

plt.gca().xaxis.set_major_formatter(date_format)

# plt.legend()
plt.title('Dates')
plt.xlabel('Day')
plt.ylabel('Mount')



plt.tight_layout()
# plt.savefig('myPlot.png')
# fig1.savefig('myPlot1.png')
# fig2.savefig('myPlot2.png')

plt.show()

# # colors:
# # blue = #008fd5
# # red = #fc4f30
# # yellow = #e5ae37
# # green = #6d904f

```
