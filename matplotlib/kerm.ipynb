```python
df = pd.read_csv('survey_results_public.csv')
pd.set_option('max_column', 85)
df
```

```python
age_group = df.groupby('Age1stCode')
x = age_group['ConvertedCompYearly'].median()
df2 = pd.DataFrame(x).reset_index()
df2.rename(columns={'ConvertedCompYearly' : 'All Dev Median Salary'}, inplace=True)
df2
```

```python
# df2.iloc[8], df2.iloc[0]= df2.iloc[0], df2.iloc[8]
def move_up(df, index, destination):
    while True:
        df.iloc[index], df.iloc[index - 1] = df.iloc[index - 1], df.iloc[index]
        index -= 1
        if index == destination:
            break
    
    
move_up(df2, 8, 0)
df2
```
