```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

plt.style.use('fivethirtyeight')

df = pd.read_csv('/Users/mhd/Desktop/matplotlib/forscatterPlot.csv')

# x = np.random.randint(0, 15, size=20)
# y = np.random.randint(0, 15, size=20)
# ration = np.random.randint(1, 10, size=20)
# size = np.random.randint(1, 300, size=20)


yearsCode = df['YearsCode'].tolist()
salary = df['ConvertedCompYearly'].tolist()
size = np.random.randint(1, 300, size=len(salary))
color = np.random.randint(1, 300, size=len(salary))


# s for size
# c for color
# marker
# cmap Greens Reds and need cbar for explain
plt.scatter(yearsCode, salary, s=size, c=color, cmap='summer', edgecolors='black', linewidths=1, alpha=0.75)

plt.xscale('log')
plt.yscale('log')

cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.title('Dev Salary')
plt.tight_layout()
# # plt.grid()
plt.ylabel('Salary')
plt.xlabel('Years Coding')
# plt.xticks(rotation=-45)
# plt.legend(loc=(0.05, 0.68))

# # plt.savefig('myPlot.png')
plt.show()

# # colors:
# # blue = #008fd5
# # red = #fc4f30
# # yellow = #e5ae37
# # green = #6d904f
```
