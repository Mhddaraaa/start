```python
from matplotlib import pyplot as plt
import pandas as pd

plt.style.use('fivethirtyeight')

df = pd.read_csv('/Users/mhd/Desktop/matplotlib/salary.csv')

ages = df['Age1stCode']
all_dev = df['All Dev Median Salary']
py_dev = df['Py Dev Median Salary']
js_dev = df['Js Dev Median Salary']

py_dev = [44000.0, 59454.0, 45379.0, 47000.5, 75000.0, 90000.0, 77831.0, 85395.0, 86671.5]

median_salary = df['All Dev Median Salary'].median()

plt.plot(ages, all_dev, linewidth=2, linestyle='--', label='All dev')
plt.plot(ages, py_dev, linewidth=2, label='py dev')

# plt.fill_between(ages, py_dev, median_salary, alpha=0.25)

# plt.fill_between(ages, py_dev, median_salary, where=(py_dev > median_salary), interpolate=True, color='green',
#                  alpha=0.25)
# plt.fill_between(ages, py_dev, median_salary, where=(py_dev <= median_salary), interpolate=True, color='red',
#                  alpha=0.25)

plt.fill_between(ages, py_dev, all_dev, where=(py_dev > all_dev), interpolate=True, color='green',
                 alpha=0.25, label = 'Above all dev salary')
plt.fill_between(ages, py_dev, all_dev, where=(py_dev <= all_dev), interpolate=True, color='red',
                 alpha=0.25, label = 'Below all dev salary')

plt.title('Dev Salary')
plt.tight_layout()
# # plt.grid()
plt.ylabel('Median Salary')
plt.xlabel('Ages')
plt.xticks(rotation=-45)
plt.legend(loc=(0.05, 0.68))  # loc = (right prcnt, up prcnt)

# # plt.savefig('myPlot.png')
plt.show()

# # colors:
# # blue = #008fd5
# # red = #fc4f30
# # yellow = #e5ae37
# # green = #6d904f
```
