```python
import pandas as pd
```

```python
df = pd.read_html('https://www.faa.gov/air_traffic/publications/atpubs/cnt_html/appendix_a.html')
```

```python
df[0]
```

```python
df_new1 = df[0][['STATE(TERRITORY)', 'STATE(TERRITORY).1']]
df_new2 = df[0][['STATE(TERRITORY).2', 'STATE(TERRITORY).3']]
df_new3 = df[0][['STATE(TERRITORY).4', 'STATE(TERRITORY).5']]
```

```python
df_new1.rename(columns={'STATE(TERRITORY)':'State', 'STATE(TERRITORY).1':'Abbreviation'}, inplace=True)
df_new2.rename(columns={'STATE(TERRITORY).2':'State', 'STATE(TERRITORY).3':'Abbreviation'}, inplace=True)
df_new3.rename(columns={'STATE(TERRITORY).4':'State', 'STATE(TERRITORY).5':'Abbreviation'}, inplace=True)
```

```python
dff = pd.concat([df_new1, df_new2, df_new3])
dff.set_index('State')
```
