```python
import pandas as pd
```

```python
df = pd.read_csv('vgsales.csv')

people = {
    'first name': ['mohammad', 'mostafa', 'benyamin',],
    'last name': ['daraee', 'dana', 'dana'],
    'email':['test@gmail.com', 'test3@gmail.com', 'test4@gmail.com']
}

df2 = pd.DataFrame(people)
df
```

```python
df['Platform'].value_counts()
```

```python
df2
```

```python
filt = (df2['last name'] == 'dana')
df2[filt].index
```

```python
df2[filt]
```

```python
df2.loc[filt, ['first name', 'last name']]
```

```python
df2[(df2['last name'] == 'dana') & (df2['first name'] == 'mohammad')]
```

```python
df2[(df2['last name'] == 'dana') | (df2['first name'] == 'mohammad')]
```

```python
df2[-filt]  #oposite of filter
```
