```python
# loading the image
img = tf.keras.utils.load_img('/content/frog.jpeg')
# resize the image 224x224 square image as vgg16 accept
img = img.resize((224, 224))
# convert image to array
img_array = tf.keras.utils.img_to_array(img)
# expand image dimention to be as (batch, height, width, channels)
img_array = np.expand_dims(img_array, axis=0)
```

```python
# preprocess the inpute image
img_array = tf.keras.applications.imagenet_utils.preprocess_input(img_array)
```

```python
# load pretrained VGG16 model
pretrained_model = tf.keras.applications.VGG16(weights='imagenet')
```

```python
# predict using model.predict
prediction = pretrained_model.predict(img_array)
```

```python
# now decode the prediction
actual_prediction = tf.keras.applications.imagenet_utils.decode_predictions(prediction)
actual_prediction[0][0]
```

```python
pretrained_model_VGG19 = tf.keras.applications.VGG19(weights='imagenet')
prediction = pretrained_model_VGG19.predict(img_array)
actual_prediction = tf.keras.applications.imagenet_utils.decode_predictions(prediction)
actual_prediction[0][0]
```

```python
pretrained_model_Resnet50 = tf.keras.applications.ResNet50(weights='imagenet')
prediction = pretrained_model_Resnet50.predict(img_array)
actual_prediction = tf.keras.applications.imagenet_utils.decode_predictions(prediction)
actual_prediction[0][0]
```
